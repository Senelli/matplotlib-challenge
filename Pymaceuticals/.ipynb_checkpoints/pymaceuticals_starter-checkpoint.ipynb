```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
mouse_study_combined_df = pd.merge(study_results, mouse_metadata, on="Mouse ID", how="left")

# Display the data table for preview
mouse_study_combined_df.head()
```

```python
# Checking the number of mice.
mice_count = len(mouse_study_combined_df["Mouse ID"].unique())
mice_count
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicated_mice = mouse_study_combined_df.loc[mouse_study_combined_df.duplicated(subset=["Mouse ID", "Timepoint"]),
                                          "Mouse ID"].unique() 
duplicated_mice
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicated_mice_df = mouse_study_combined_df.loc[mouse_study_combined_df["Mouse ID"] == "g989"]
duplicated_mice_df
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_mouse_study_combined_df = mouse_study_combined_df[mouse_study_combined_df['Mouse ID'] != 'g989']
clean_mouse_study_combined_df
```

```python
# Checking the number of mice in the clean DataFrame.
total_mice = len(clean_mouse_study_combined_df["Mouse ID"].unique())
total_mice
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
mean_tumor = clean_mouse_study_combined_df.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].mean()
median_tumor = clean_mouse_study_combined_df.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].median()
var_tumor = clean_mouse_study_combined_df.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].var()
std_tumor = clean_mouse_study_combined_df.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].std()
sem_tumor = clean_mouse_study_combined_df.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].sem()
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.
drug_tumor_data_df = pd.DataFrame({"Mean Tumor Volume": mean_tumor,
                       "Median Tumor Volume": median_tumor,
                       "Tumor Volume Variance": var_tumor,
                       "Tumor Volume Std. Dev.": std_tumor,
                       "Tumor Volume Std. Err.": sem_tumor})
drug_tumor_data_df
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line
drug_tumor_data_df = clean_mouse_study_combined_df.groupby(["Drug Regimen"])[["Tumor Volume (mm3)"]].agg(["mean", "median", "var", "std", "sem"])
drug_tumor_data_df
```

## Bar and Pie Charts

```python
# Generate a bar plot showing the total number of rows (Mouse ID/Timepoints) for each drug regimen using Pandas.
number_trials_df = clean_mouse_study_combined_df.groupby(["Drug Regimen"])["Timepoint"].count()
number_trials_df.sort_values(ascending=False).plot(kind="bar")
plt.ylabel("# of Observed Mouse Timepoints")
plt.show()
```

```python
# Generate a bar plot showing the total number of rows (Mouse ID/Timepoints) for each drug regimen using pyplot.
sorted_number_trials_df = number_trials_df.sort_values(ascending=False)
x_axis = sorted_number_trials_df.index
y_axis = sorted_number_trials_df.sort_values(ascending=False)
plt.bar(x_axis, y_axis)
plt.xlabel("Drug Regimen")
plt.ylabel("# of Observed Mouse Timepoints")
plt.xticks(rotation="vertical")
plt.show()
```

```python
# Generate a pie plot showing the distribution of female versus male mice using Pandas
mice_count = clean_mouse_study_combined_df.groupby(["Mouse ID", "Sex"]).size()
mice_count_sex = mice_count.groupby("Sex").count()
mice_count_sex
mice_count_percent = (mice_count_sex / mice_count_sex.sum()) * 100
mice_count_percent.plot.pie(startangle= 180, autopct = "%1.1f%%")
plt.title("Male vs Female Mouse Population")
plt.ylabel("Sex")
plt.show()
```

```python
# Generate a pie plot showing the distribution of female versus male mice using pyplot
percentages = []
for percent in mice_count_percent:
    percentages.append(percent)
labels = clean_mouse_study_combined_df["Sex"].unique()
plt.pie(percentages, labels=labels, autopct="%1.1f%%", startangle= 180)
plt.title("Male vs Female Mouse Population")
plt.ylabel("Sex")
plt.show()
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
max_timepoint = clean_mouse_study_combined_df.groupby(["Mouse ID"])["Timepoint"].max()
max_timepoint_df = pd.DataFrame(max_timepoint)

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
merged_timepoint_df = pd.merge(max_timepoint_df, clean_mouse_study_combined_df, on=["Mouse ID", "Timepoint"], how='left')
merged_timepoint_df
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol_data = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for treatment in treatments:
    # Locate the rows which contain mice on each drug and get the tumor volumes
    drug_to_tumor_vol = merged_timepoint_df.loc[merged_timepoint_df["Drug Regimen"] == treatment, "Tumor Volume (mm3)"]
    tumor_vol_data.append(drug_to_tumor_vol)
    
    # Determine outliers using upper and lower bounds
    quartiles = drug_to_tumor_vol.quantile([0.25, 0.5, 0.75])
    iqr = quartiles[0.75] - quartiles[0.25]

    lower_bound = quartiles[0.25] - 1.5 * iqr
    upper_bound =  quartiles[0.75] + 1.5 * iqr

    outliers = drug_to_tumor_vol[(drug_to_tumor_vol < lower_bound) | (drug_to_tumor_vol > upper_bound)]
    print(f"Potential outliers for {treatment}: {outliers}")
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
filter_prop = dict(marker="o", markerfacecolor='r')
plt.boxplot(tumor_vol_data, flierprops=filter_prop)
plt.xticks([1, 2, 3, 4], [f"{i}" for i in treatments])
plt.ylabel("Final Tumor Volume (mm3)")
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomulin
line_plot_df = clean_mouse_study_combined_df.loc[clean_mouse_study_combined_df['Mouse ID'] == 'l509',:]
x = line_plot_df["Timepoint"]
y = line_plot_df["Tumor Volume (mm3)"]
plt.plot(x, y, linewidth=2, markersize= 10)
plt.title("Capomulin treatment of mouse l509")
plt.ylabel("Tumor Volume (mm3)")
plt.xlabel("Timepoint (days)")
plt.show()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
avg_Capomulin = clean_mouse_study_combined_df.loc[clean_mouse_study_combined_df['Drug Regimen'] == 'Capomulin']

# Find average tumor volume for each mouse
avg_tumor_vol_df = pd.DataFrame(avg_Capomulin.groupby('Mouse ID')['Tumor Volume (mm3)'].mean().sort_values()).reset_index().rename(columns={'Tumor Volume (mm3)':'Average'})
scatter_plot_df = pd.merge(avg_Capomulin, avg_tumor_vol_df, on='Mouse ID')
final_scatter_plot_df = scatter_plot_df[['Weight (g)', 'Average']].drop_duplicates()

x = final_scatter_plot_df['Weight (g)']
y = final_scatter_plot_df['Average']
```

```python
plt.scatter(x,y)
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.title("Average Tumor Volume by Weight")
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen
corr = st.pearsonr(x,y)
print(f"The correlation between mouse weight and the average tumor volume is {round(corr[0],2)}")

# Perform a linear regression on mouse weight vs average tumor volume
(slope, intercept, rvalue, pvalue, stderr) = st.linregress(x, y)
regress_values = x * slope + intercept
plt.scatter(x,y)
plt.plot(x,regress_values,"r-")
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.title('Average Tumor Volume by Weight')
plt.show()
```
