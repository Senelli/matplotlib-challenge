```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
mouse_study_combined_df = pd.merge(study_results, mouse_metadata, on="Mouse ID", how="left")

# Display the data table for preview
mouse_study_combined_df.head()
```

```python
# Checking the number of mice.
mice_count = len(mouse_study_combined_df["Mouse ID"].unique())
mice_count
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicated_mice = mouse_study_combined_df.loc[mouse_study_combined_df.duplicated(subset=["Mouse ID", "Timepoint"]),
                                          "Mouse ID"].unique() 
duplicated_mice
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicated_mice_df = mouse_study_combined_df.loc[mouse_study_combined_df["Mouse ID"] == "g989"]
duplicated_mice_df
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
#clean_mouse_study_combined = mouse_study_combined_df.drop_duplicates(subset=['Mouse ID', 'Timepoint'], keep=False)
#clean_mouse_study_combined_df = mouse_study_combined_df.drop_duplicates(subset=['Mouse ID'])
clean_mouse_study_combined_df = mouse_study_combined_df[mouse_study_combined_df['Mouse ID'].isin(duplicated_mice)==False]
clean_mouse_study_combined_df
```

```python
# Checking the number of mice in the clean DataFrame.
total_mice = len(clean_mouse_study_combined_df["Mouse ID"].unique())
total_mice
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
mean_tumor = clean_mouse_study_combined_df.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].mean()
median_tumor = clean_mouse_study_combined_df.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].median()
var_tumor = clean_mouse_study_combined_df.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].var()
std_tumor = clean_mouse_study_combined_df.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].std()
sem_tumor = clean_mouse_study_combined_df.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].sem()
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.
drug_tumor_data_df = pd.DataFrame({"Mean Tumor Volume": mean_tumor,
                       "Median Tumor Volume": median_tumor,
                       "Tumor Volume Variance": var_tumor,
                       "Tumor Volume Std. Dev.": std_tumor,
                       "Tumor Volume Std. Err.": sem_tumor})
drug_tumor_data_df
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line
drug_tumor_data_df = clean_mouse_study_combined_df.groupby(["Drug Regimen"])[["Tumor Volume (mm3)"]].agg(["mean", "median", "var", "std", "sem"])
drug_tumor_data_df
```

## Bar and Pie Charts

```python
# Generate a bar plot showing the total number of rows (Mouse ID/Timepoints) for each drug regimen using Pandas.
number_trials_df = clean_mouse_study_combined_df.groupby(["Drug Regimen"])["Timepoint"].count()
number_trials_df
number_trials_df.plot(kind="bar")
plt.ylabel("# of Observed Mouse Timepoints")
plt.show()
```

```python
# Generate a bar plot showing the total number of rows (Mouse ID/Timepoints) for each drug regimen using pyplot.
x_axis = number_trials_df.index.values
y_axis = number_trials_df.values
plt.bar(x_axis, y_axis, alpha=0.8, align="center")
plt.xlabel("Drug Regimen")
plt.ylabel("# of Observed Mouse Timepoints")
plt.xticks(rotation="vertical")
plt.show()
```
